<h2>Splitting the data where there is no profit which means only loss</h2>

```python
weakdf=df[df['Profit']<0]
```

```python
weakdf.head()
```

```python
weakdf.shape
```

```python
weakdf.describe()
```

<h2>Bar plot on What category has the higher Loss</h2>

```python
plt.rcParams['figure.figsize']=(15,3)
plt.bar(weakdf['Category'],weakdf['Profit'])
plt.rcParams.update({'font.size':10})
plt.xlabel('Category')
plt.ylabel('Profit')
```

<h2>As we can see Technology plays a huge role in loss as the loss is greater in technology.</h2>

```python
region=region = df.groupby('Category')['Discount'].sum().sort_values(ascending=False).head(10)
region =region.reset_index()
print(region)
```

```python
region = df.groupby('Category')['Discount'].sum().sort_values(ascending=False).head(10)
region = region.reset_index()
plt.pie(region['Discount'], labels=region['Category'], autopct='%1.1f%%')
plt.title('Categories by Discount')
plt.show()
```

<h2>As we can see Discounts are given more to office supplies and technology has low discounts.as it becomes a WEAK AREA.To over come this more discounts should be given on technology so as to decrease the loss and increase sales</h2>

<h2>plots on sales ,profit and discount</h2>

```python
plt.scatter(weakdf['Profit'], weakdf['Sales'])
plt.xlabel("Profit")
plt.ylabel('Sales')
plt.title('Profit vs Sales')
plt.show()
plt.scatter(weakdf['Discount'], weakdf['Sales'])
plt.xlabel('Discount')
plt.ylabel('Sales')
plt.title('Discount vs Sales')
plt.show()
plt.scatter(weakdf['Profit'], weakdf['Discount'])
plt.xlabel("Profit")
plt.ylabel('Discount')
plt.title("Profit vs Discount")
plt.show()

```

<h2>From Profit vs Sales plot we come to know that not everytime high sales gives high profit<br>From Discount vs Sales plot we come to know that discounts should be given between 40%-50% for better sales<br>From Profit vs Discount plot we come to know that 25% to 35% discount gives less loss.</h2> 

```python
tdf = weakdf[weakdf["Category"] == "Technology"]
gdf = tdf.groupby("Region").size().reset_index(name='Technology count').sort_values('Technology count', ascending=False)
gdf
```

```python
# Assuming gdf contains the DataFrame with technology count by region
# Set the index of the DataFrame to the "Region" column
gdf.set_index('Region', inplace=True)
# Generate a pie chart from the DataFrame
gdf.plot(kind='pie', y='Technology count', legend=False, autopct='%1.1f%%', startangle=90)
# Set the title of the pie chart
plt.title('Technology count by region')
# Show the pie chart
plt.show()
```

<h2>Technology purchase count by Regions shows that South Region has the least Technology count.Manufacturing of less technology goods at south region can be decreased to reduce loss at the southern region </h2>

```python
# Assuming tdf contains the DataFrame with the technology category
technology_subcat_counts = weakdf[weakdf["Category"] == "Technology"]["Sub-Category"].value_counts()
print(technology_subcat_counts)
```

```python
technology_subcat_counts.plot(kind='pie', legend=False, autopct='%1.1f%%', startangle=90)
plt.title('Technology count by sub-category')
plt.show()
```

<h2>Sub-category Machines have less count either machine manufacturing can be reduced or more discounts should be given to increase sales and decrease the loss.</h2>

<h2>Office supplies are next large contributors to the loss</h2>

```python
import pandas as pd

osregions = weakdf[weakdf["Category"] == "Office Supplies"].groupby("Region").size().reset_index(name='Count').sort_values(by='Count', ascending=False)

osregions
```
